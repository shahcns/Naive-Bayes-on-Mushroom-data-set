# 3. <font color='Red'> Feature Scaling </font>

# 4. <font color='Red'> Modeling  </font>   
## a. Splitting Data set into independent variable and Dependent variable.

```python
X=df1.iloc[:,1:].values
y=df1.iloc[:,[0]].values
X
```

**Note:**<br>
Here we took the whole variables except the "Class" column because the we can see that in given above link, which was the study about the Mushroom, in that it is clear that all preticulars of Head, Stem/Stalk and root has much influence in order to determine wheather it is edible or not.

## b. Scaling the Splitted Data set.

```python
X=StandardScaler().fit_transform(X)
X
```

## c. Further splitted the data set into Training set and Testing data.

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)
print('The no of rows and columns of training set of independent variables is:', X_train.shape)
```

## d. Fitting the different flavours of NB.
### <font color='tomato'>I Gaussing Naive Bayes </font>

```python
fit1=GaussianNB().fit(X_train,y_train,)
y_pred_G=fit1.predict(X_test)
y_pred_G
warnings.filterwarnings('ignore')
```

```python
c1=pd.DataFrame(y_test,y_pred_G)
c1.reset_index(inplace=True)
c1.rename(columns={'index':'Actual',0:'Predicted'},inplace=True)
c1.head()
```

```python
print('The Confusion Matrix of Gaussian Naive Bayes is :',confusion_matrix(y_test,y_pred_G))
```

Here we can see that among 2438, 2266 were correctly predicted and 172 wrongly predicted.

```python
print('The Accuracy score of Gaussian Naive Bayes is :',accuracy_score(y_test,y_pred_G))
```

### <font color='tomato'>II Bernoulli Naive Bayes </font>

```python
fit2=BernoulliNB().fit(X_train,y_train,)
y_pred_B=fit2.predict(X_test)
y_pred_B
```

```python
c2=pd.DataFrame(y_test,y_pred_B)
c2.reset_index(inplace=True)
c2.rename(columns={'index':'Actual',0:'Predicted'},inplace=True)
c2.head()
```

```python
print('The Confusion Matrix of Bernoulli Naive Bayes is :',confusion_matrix(y_test,y_pred_B))
```

Here we can see that among 2438, 2191 were correctly predicted and 247 wrongly predicted.

```python
print('The Accuracy score of Gaussian Naive Bayes is :',accuracy_score(y_test,y_pred_B))
```

```python
X1=df1.iloc[:,1:].values
y1=df1.iloc[:,[0]].values
```

```python
X1_train,X1_test,y1_train,y1_test=train_test_split(X1,y1,random_state=12,test_size=0.3)
```

```python
c3=pd.DataFrame(y_test,y_pred_M)
c3.reset_index(inplace=True)
c3.rename(columns={'index':'Actual',0:'Predicted'},inplace=True)
c3.head()
```

```python
print('The Confusion Matrix of Multinomial Naive Bayes is :',confusion_matrix(y1_test,y_pred_M))
```

Like we saw in the above table,there may have more wrongful predictions occurs when we use the Multinomial Naive Bayes. Using the confusion matrix we can say that out 2438, the correctly predicted is 1968 and wrongly predicted is 470.

```python
print('The Accuracy score of Multinomial Naive Bayes is :',accuracy_score(y1_test,y_pred_M))
```
